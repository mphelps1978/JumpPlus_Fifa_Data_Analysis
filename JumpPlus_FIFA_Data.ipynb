```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None )
pd.set_option('display.max_rows', None )
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
ds = pd.read_csv('./data/players_20.csv')
df = pd.DataFrame(ds)
```

```python
## Some global functions that will reduce the redundancy of some code (Keep it DRY) ##

# get a list of players from any column
def get_players(c, v):
    return df.loc[df[c] == v ]
```

```python
# Req. 1 - Show top 5 records
df.head()
```

```python
# Req. 2 - Display Each Column Name
list(df.columns)
```

```python
# Req. #3 - Show the number of rows and columns in the dataset
print("The Number of Rows: ", len(df))
print("The Number of Columns: ", len(df.columns))
```

```python
# Req #4 Show the number of players and their countries
print(df.groupby(['nationality'])['nationality'].count())
```

```python
# Req #5 - Show top 10 countries and their number of players
top10 = df.groupby(['nationality'])['nationality'].count().head(10).sort_values(ascending = False)
top10
```

```python
# Another way of doing 5, sorting the whole list, as opposed to just the top10
top5 = df.groupby(['nationality'])['nationality'].count().sort_values(ascending = False).head()
print(top5)
top5.plot.bar(color='green', title="Top 5 Nationalities with the most players");
```

```python
# Top 5 players short name and wages
top5_players_wages = df[['short_name', 'wage_eur']].head()
top5_players_wages
```

```python
#8
highest_salaries = df.sort_values(by='wage_eur', ascending=False)[['short_name', 'wage_eur']].head()
print(highest_salaries)
highest_salaries.plot.bar(x='short_name', color='red', title="Top 5 Highest Paid Salaries");
```
